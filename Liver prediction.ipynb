Age: The age of the individual, represented as a numerical value. This feature captures the age of the patient, which can be an important factor in liver-related conditions.

Gender: The gender of the individual, represented as a categorical variable (e.g., Male or Female). Gender may play a role in differentiating liver-related diseases, as some conditions are more prevalent in specific genders.

Total_Bilirubin: Total bilirubin is a liver enzyme measured in the blood. Elevated levels of total bilirubin can indicate potential liver dysfunction or diseases.

Direct_Bilirubin: Direct bilirubin is a fraction of the total bilirubin and can also indicate liver function or abnormalities.

Alkaline_Phosphatase: Alkaline phosphatase is another liver enzyme, and its levels in the blood can provide insights into liver health.

Alamine_Aminotransferase: This is an enzyme found in the liver, and elevated levels may indicate liver damage or disease.

Aspartate_Aminotransferase: Like Alamine Aminotransferase, this enzyme is also found in the liver, and its levels can be indicative of liver health.

Total_Protiens: Total proteins in the blood, including albumin and globulins, which are crucial for various bodily functions. Abnormal levels can point to liver-related issues.

Albumin: Albumin is a protein synthesized by the liver. Its levels in the blood can reflect liver function.

Albumin_and_Globulin_Ratio: The ratio of albumin to globulin levels, which can provide additional insights into liver health.

Dataset: This is the target variable and typically represents the presence or absence of a liver disease or condition. It is a binary variable (0 or 1), with 0 indicating no liver disease and 1 indicating the presence of a liver-related condition.

# scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
se=StandardScaler()
```

```python
Xtrain=se.fit_transform(Xtrain)
Xtest=se.fit_transform(Xtest)
```

# modeling`

```python
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from xgboost import  XGBClassifier

from sklearn.metrics import accuracy_score,classification_report
from sklearn.metrics import roc_auc_score,roc_curve
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
lr=LogisticRegression()
dt=DecisionTreeClassifier()
ra=RandomForestClassifier()
ad=AdaBoostClassifier()
svm=SVC(probability=True)
gau=GaussianNB()
bag=BaggingClassifier()
Gr=GradientBoostingClassifier()
xg= XGBClassifier()
```

```python
Training_score= []
Testing_score= []
def model_building(model):
    model.fit(Xtrain, ytrain)
    ytrain_pred= model.predict(Xtrain)
    ytest_pred= model.predict(Xtest)
    a= accuracy_score(ytrain, ytrain_pred)
    b= accuracy_score(ytest, ytest_pred)
    Training_score.append(a)
    Testing_score.append(b)
    print(model)
    print("Train Data\n", accuracy_score(ytrain,ytrain_pred))
    print("Test Data\n", accuracy_score(ytest,ytest_pred))
```

```python
def roc_auc_curve(model):
    print("ROC-AUC CURVE:")
    ytest_prob= model.predict_proba(Xtest)
    fpr,tpr,thresh=roc_curve(ytest,ytest_prob[:,1])
    print('roc_score')
    roc_score=roc_auc_score(ytest,ytest_prob[:,1])
    
    sns.lineplot(x=fpr,y=tpr)
    sns.lineplot(x=[0.0,1.0],y=[0.0,1.0],color="red",linestyle="--")
    plt.xlabel("FPR")
    plt.ylabel("TPR")
    plt.title("ROC-AUC Curve")
    plt.show()
    print(roc_score)
```

```python
from sklearn.metrics import log_loss

def calculate_logloss(model,ytest):
    print('log_loss:')
    ytest_prob= model.predict_proba(Xtest)

    logloss = log_loss(ytest, ytest_prob[:,-1])
    return logloss
```

```python
model_building(knn)
roc_auc_curve(knn)
calculate_logloss(knn,ytest)
```

# Random forest

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_forest_params = {
    'n_estimators': [25,50,75,100],
    'max_depth': [2, 3, 5, 10, 20],
    'min_samples_leaf': [5, 10, 20, 50, 100],
    'min_samples_split': [2, 5, 10],
    'criterion': ["gini", "entropy"],
    'max_features': ['auto', 'sqrt'],
    'bootstrap': [True, False]
}
```

```python
ra_reg=RandomizedSearchCV(ra,param_distributions=random_forest_params,random_state=42,scoring='accuracy',cv=5,n_jobs=-1)
```
